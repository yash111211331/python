```python
# 1. class and object
class Animal:
    def __init__(self, name):
        self.name = name  # attribute
    def speak(self):
        return f'{self.name} makes a sound.'
    # create object
dog = Animal('Dog')
print(dog.speak())
cat = Animal('Cat')
print(cat.speak())
cow = Animal('Cow')
print(cow.speak())
```

```python
# 2. inheritance
class Dog(Animal):                    # dog inherit from animal
    def speak(self):
        return f'{self.name} barks!'
class Cat(Animal):                     # cat inherit from animal
    def speak(self):
        return f'{self.name} meows!'
class Cow(Animal):                   # cow inherit from animal
    def speak(self):
        return f'{self.name} moo!'
# object of child class
dog = Dog('Dog')
cat = Cat('Cat')
cow = Cow('Cow')
print(dog.speak())
print(cat.speak())
print(cow.speak())
```

```python
# 3. polymorphism
animals = [Dog('Dog'), Cat('Cat'), Cow('Cow'), Animal('generic Animal')]
for a in animals:
    # same method name 'speak' behave differently depending on object type
    print(a.speak())
```

```python
# 4. Encapsulation
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # private attribute (name mangling)
    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited {amount}, New Balance: {self.__balance}")
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}, New Balance: {self.__balance}")
        else:
            print("insufficient funds")
    def get_balance(self):
        return self.__balance
# test encapsulation
account = BankAccount('Charlie Kirk {Ore wa shinda}', 50000)
account.deposit(50000)
account.withdraw(100000)
print("final balance --> ",account.get_balance(),' // owner name --> ',account.owner)
print("--------------------------------------------------------------------------------------------------------")
account = BankAccount('Erica Kirk',499)
account.deposit(100000)
account.withdraw(1499)
print("final balance --> ",account.get_balance(),' // owner name --> ',account.owner)
```

```python
# 5. Abstraction
from abc import ABC, abstractmethod
class Vecihle(ABC):          # abstract base class
    @abstractmethod
    def start(self):
        pass

class Car(Vecihle):
    def start(self):
        return "car engine start with a button"

class Bike(Vecihle):
    def start(self):
        return "bike engine start with a key"

# abstract class cannot be instantiated directly
car = Car()
bike = Bike()
print(car.start())
print(bike.start())
```
